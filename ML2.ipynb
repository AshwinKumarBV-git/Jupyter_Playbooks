```python
#EXPERIMENT 2
#Read a dataset and use an example based(ripper/CN2) method to generate a set of classification rules. 
#Apply the FOIL algorithm to learn first order rules for predicting.

#Using CN2

import Orange
from Orange.classification import CN2Learner

def main():
    data=Orange.data.Table("iris")
    cn2=CN2Learner()
    classifier=cn2(data)
    print("Learned CN2 Rules:")
    for rule in classifier.rule_list:
        print(rule)
    print("\n Predictions on first 5 instances:")
    for instance in data[:5]:
        pred=classifier(instance)
        true_class=instance.get_class()
        true_label=data.domain.class_var.values[int(true_class)]
        pred_label=data.domain.class_var.values[int(pred)]
        print(f"True:{true_label},Predicted:{pred_label}")

if __name__=="__main__":
    main()
                        
```
