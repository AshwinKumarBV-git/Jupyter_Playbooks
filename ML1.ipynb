```python
#EXPERIMENT 1
#FIND-S ALGORITHM

import pandas as pd
import numpy as np

def find_s(examples):
    examples=np.array(examples)
    num_attributes=(examples.shape[1]-1)
    hypothesis=['ϕ'] *num_attributes
    for example in examples:
        if example[-1].lower()=='yes':
            for i in range(len(hypothesis)):
                if hypothesis[i]== 'ϕ': 
                    hypothesis[i]=example[i]
                    
                elif hypothesis[i]!=example[i]:
                    hypothesis[i]='?'
    return hypothesis
data=pd.read_csv("tennis_dataset.csv")

final_hypothesis=find_s(data)
print("Final hypothesis:",final_hypothesis)
```

```python
#CANDIDATE ELIMINATION ALGORITHM
import pandas as pd
import numpy as np

def candidate_elimination(data):
    data=np.array(data)
    X,y=data[:, :-1],data[:, -1]
    S=list(X[0])if y[0]=='Yes' else ['ϕ']*X.shape[1]
    G=[['?']*X.shape[1]]

    for i,example in enumerate(X):
        if y[i]=='Yes':
            for j in range(len(S)):
                if S[j]=='φ':
                    S[j]=example[j]
                elif S[j]!=example[j]:
                    S[j]='?'
            G=[g for g in G if all(g[j]=='?' or S[j]=='?' or g[j]==S[j] for j in range(len(S)))]
        else:
            G_new = []
            for g in G:
                for j in range(len(g)):
                    if g[j]=='?':
                        if S[j]!=example[j]:
                            new_g=g.copy()
                            new_g[j]=S[j]
                            if new_g not in G_new:
                                G_new.append(new_g)
            G=G_new
    return S,G
data=pd.read_csv("tennis_dataset.csv")
S,G=candidate_elimination(data)

print("Final S(Specific):",S)
print("Final G(general):",G)
```
