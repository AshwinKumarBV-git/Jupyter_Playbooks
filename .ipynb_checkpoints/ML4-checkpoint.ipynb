```python
# EXPERIMENT 4 
# Read an unsupervised dataset and group the dataset based on similarity using K-Means clustering

from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

data=load_iris().data
scaled=StandardScaler().fit_transform(data)
labels=KMeans(n_clusters=3,random_state=42).fit_predict(scaled)
print(pd.DataFrame(data).assign(Cluster=labels).head())
```
