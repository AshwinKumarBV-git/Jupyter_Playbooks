```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.utils import plot_model
```

```python


# XOR input and output
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])

```

```python
# Define model using Input layer
model = Sequential([
    Input(shape=(2,)),            # Input layer with 2 features
    Dense(2, activation='relu'),  # Hidden layer with 2 neurons
    Dense(1, activation='sigmoid')# Output layer
])

```

```python
# Compile and train
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X, y, epochs=10, verbose=0)


```

```python
# Evaluate
_, accuracy = model.evaluate(X, y)
print(f"Accuracy: {accuracy * 100:.2f}%")

# Predict
predictions = np.round(model.predict(X)).astype(int)
print("Predictions:")
for i in range(len(X)):
    print(f"Input: {X[i]} => Predicted Output: {predictions[i][0]}, Actual Output: {y[i][0]}")

# Plot architecture
plot_model(model, to_file='xor_model.png', show_shapes=True, show_layer_names=True)
print("Model diagram saved as 'xor_model.png'")
```

```python
#recursion on subsequences , take or not take , making choices , printing subsequences
arr = [3 , 2 , 1]
res = []
def subc(index):
    if index == len(arr):
        print(res)
        return
    res.append(arr[index]) #taking
    subc(index + 1)
    res.pop() #not taking
    subc(index + 1)
subc(0)
```

```python
arr = [3 , 2 , 1]
res = []
k = 3
suma = 0

def subcs(index,suma):
    if index == len(arr):
        if suma == k:
            print(res)
        return
        
    suma += arr[index]
    res.append(arr[index]) #taking 
    subcs(index + 1,suma)
    
    suma -= arr[index]
    res.pop() #not taking
    subcs(index + 1,suma)
    
subcs(0,suma)
```

```python
#printing only one subsequence
arr = [3 , 2 , 1]
res = []
k = 3
suma = 0

def subcs(index,suma):
    #base condition
    if index == len(arr):
        if suma == k:
            print(res)
            return True
        return False
        
    suma += arr[index]
    res.append(arr[index]) #taking 
    if subcs(index + 1,suma) == True:
        return True #no more extra recursion calls
    
    suma -= arr[index]
    res.pop() #not taking
    if subcs(index + 1,suma) == True:
        return True #no more extra calls
    return False
    
subcs(0,suma)
```

```python
#printing only count
arr = [3 , 2 , 1]
res = []
k = 3
suma = 0

def subcs(index,suma):
    #base condition
    if index == len(arr):
        if suma == k:
            print(res)
            return 1
        return 0
        
    suma += arr[index]
    res.append(arr[index]) #taking 
    l = subcs(index + 1,suma)
    
    suma -= arr[index]
    res.pop() #not taking
    r = subcs(index + 1,suma) 
    return l+r
    
print("count =" ,subcs(0,suma))
```

```python
nums = [3 , 2 , 1]
res = []
n = len(nums)
def backtrack(index , arr):
    if index == n:
        res.append(sum(arr))
        return
    arr.append(nums[index])
    backtrack(index + 1 , arr)
    arr.pop()
    backtrack(index + 1 , arr)
backtrack(0, [])
print(res)
```
